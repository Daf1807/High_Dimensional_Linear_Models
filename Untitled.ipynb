* Guardas estos datos (x, y) porque serán la base para todos los experimentos.

============================================================
Ejercicio 2 - Pregunta 2 
Evaluar el desempeño del modelo polinomial con distinto número
de predictores (features) y comparar R² en train y test.
============================================================

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

1. Generar datos base (simulación con ruido)

```python
np.random.seed(42)
n = 100
x = np.linspace(0, 10, n).reshape(-1, 1)
y = 3 + 2 * x + 0.5 * x**2 - 0.05 * x**3 + np.random.normal(0, 5, size=(n, 1))
```

2. Dividir en conjunto de entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

 3. Evaluación de modelos polinomiales con loop

```python
resultados = []

# Probar polinomios desde grado 1 hasta 10
for grado in range(1, 11):
    # Crear transformador polinomial
    poly = PolynomialFeatures(degree=grado, include_bias=False)
    
    # Transformar variables predictoras
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    
    # Ajustar modelo de regresión lineal
    modelo = LinearRegression()
    modelo.fit(X_train_poly, y_train)
    
    # Predicciones
    y_train_pred = modelo.predict(X_train_poly)
    y_test_pred = modelo.predict(X_test_poly)
    
    # Calcular R² en train y test
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    # Guardar resultados
    resultados.append({"Grado": grado, "R2_train": r2_train, "R2_test": r2_test})
```

```python
# Convertir resultados en DataFrame
df_resultados = pd.DataFrame(resultados)
```

4. Mostrar resultados

```python
print("Resultados del R² según grado del polinomio:\n")
print(df_resultados)
```

 5. Graficar desempeño

```python
plt.figure(figsize=(8,5))
plt.plot(df_resultados["Grado"], df_resultados["R2_train"], marker="o", label="Train R²")
plt.plot(df_resultados["Grado"], df_resultados["R2_test"], marker="s", label="Test R²")
plt.xlabel("Grado del polinomio")
plt.ylabel("R²")
plt.title("Desempeño del modelo polinomial en Train/Test")
plt.legend()
plt.grid(True)
plt.show()
```
